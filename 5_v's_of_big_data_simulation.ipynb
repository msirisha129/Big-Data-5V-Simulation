```python
# Simulation of Big Data Characteristics (5 V's) within a Hospital Database

import pandas as pd
import numpy as np
import time
import json
import random

print("=== HOSPITAL BIG DATA SYSTEM 5V ANALYSIS ===\n")


# ----------------------------------------------------
# 1. VARIETY : Handling Different Data Formats
# ----------------------------------------------------
print("--- 1. Variety: Handling Diverse Data Formats ---")

# Structured Data (Excel / Database style)
structured_excel = pd.DataFrame({
    "patient_id": [1001, 1002],
    "age": [45, 39],
    "blood_type": ["A+", "O-"]
})

print("\nStructured Data (Excel Style):")
print(structured_excel)


# Semi Structured Data (JSON / Log style)
semi_structured_log = {
    "timestamp": "2026-02-08T21:45:02",
    "event": "Lab Result Upload",
    "status": 200
}

print("\nSemi-Structured Data (JSON Log):")
print(json.dumps(semi_structured_log, indent=2))


# Unstructured Data (Media / Files)
unstructured_data = "Files: chest_xray_1001.jpg (image), cardiac_ultrasound.mp4 (video)"

print("\nUnstructured Data (Images/Videos):")
print(unstructured_data)


# ----------------------------------------------------
# 2. VOLUME : Large Amount of Data
# ----------------------------------------------------
print("\n--- 2. Volume: Aggregating Hospital Records ---")

total_records = 1000000
print(f"Total Database Volume: {total_records} records collected")


# ----------------------------------------------------
# 3. VELOCITY : High Speed Data Generation
# ----------------------------------------------------
print("\n--- 3. Velocity: Real-Time Patient Test Stream ---")

for i in range(3):
    print(f"Incoming Test Record {i+1} generated at high speed...")
    time.sleep(0.3)

print("Streaming Processing Complete\n")


# ----------------------------------------------------
# 4. VERACITY : Data Accuracy & Trust
# ----------------------------------------------------
print("--- 4. Veracity: Data Validation ---")

raw_data_count = 3
trusted_data_count = 2
untrusted_data = raw_data_count - trusted_data_count

print(f"Accuracy Check: Filtered {trusted_data_count} trusted records and {untrusted_data} untrustworthy records\n")


# ----------------------------------------------------
# 5. VALUE : Business / Medical Insight
# ----------------------------------------------------
print("--- 5. Value: Medical Insight ---")
print("Actionable Result: Early disease detection achieved.")
print("Treatment costs reduced by 15%")
```
